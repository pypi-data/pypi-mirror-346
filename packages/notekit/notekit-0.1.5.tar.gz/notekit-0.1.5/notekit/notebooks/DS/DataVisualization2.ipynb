### Module Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Dataset Loading

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

### Box Plot

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Create boxplot
sns.boxplot(x="sex", y="age", hue="survived", data=titanic, palette="Set2")

# Set labels and title
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Age Distribution by Gender and Survival Status")
plt.legend(title="Survived", labels=["No", "Yes"])

# Show plot
plt.show()
```

# Observations:
## Age Distribution:
#### The age range is broader for males compared to females.Outliers are present, especially among older passengers.
## Survival Trends:
#### Women (female) generally have higher survival rates, especially in younger and middle age groups. Men (male) had a lower survival rate, with many younger males not surviving.
## Children's Survival:
#### Young girls and boys (ages below 10) seem to have better survival chances, possibly due to the "women and children first" policy.

# Box Plot Basics and Data Visualization - Theory with Viva Q&A

## 1. What is a Box Plot?

**Answer:**  
A box plot (also called a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It shows the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. Box plots also highlight outliers that may be present in the dataset.

---

## 2. What does a Box Plot represent?

**Answer:**  
- **Box**: Represents the interquartile range (IQR) between Q1 and Q3  
- **Whiskers**: Extend from the box to show the range of data within 1.5 times the IQR  
- **Outliers**: Points that lie outside the whiskers, indicating extreme values in the data  
- **Median Line**: The line inside the box showing the median (50th percentile) of the data

---

## 3. Why is a Box Plot useful?

**Answer:**  
A box plot helps to:  
- Visualize the central tendency, spread, and skewness of a dataset  
- Detect outliers and identify any extreme values  
- Compare distributions across multiple categories or variables  
- Get a quick summary of data distribution without needing to see all data points

---

## 4. What does "Know Your Data" mean in data visualization?

**Answer:**  
"Know Your Data" involves understanding the structure, characteristics, and relationships within your dataset. This includes:  
- Identifying types of variables (numerical, categorical)  
- Checking for missing values or duplicates  
- Analyzing the distribution of variables  
- Identifying potential correlations between variables

---

## 5. Why is knowing your data important before performing analysis?

**Answer:**  
Understanding your data helps to:  
- Select appropriate analysis techniques and visualizations  
- Detect anomalies, inconsistencies, or errors in the dataset  
- Guide hypothesis generation and testing  
- Avoid incorrect assumptions and misleading conclusions  
- Ensure that statistical assumptions (e.g., normality, homogeneity) are met

---

## 6. How do we find patterns in data using visualization?

**Answer:**  
Patterns in data can be identified by using various visualizations such as:  
- **Box plots**: For comparing distributions and detecting outliers  
- **Histograms**: To see the frequency distribution of a variable  
- **Scatter plots**: To understand relationships between two continuous variables  
- **Correlation heatmaps**: To identify relationships between multiple variables

---

## 7. Why is it important to find patterns in data?

**Answer:**  
Finding patterns is crucial because it allows us to:  
- Understand how variables are related to each other  
- Detect trends, seasonality, or anomalies  
- Form hypotheses or predictions based on observed relationships  
- Choose appropriate features for machine learning models  
- Inform decisions and derive actionable insights from data
