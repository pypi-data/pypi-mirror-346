```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'], test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)
```

```python
y_pred = naive_bayes.predict(X_test)
```

```python
accuracy = (y_pred == y_test).sum() / len(y_test)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap="crest")
```

# Naïve Bayes Classifier - Theory with Viva Q&A

## 1. What is Naïve Bayes Classifier?

**Answer:**  
Naïve Bayes is a classification algorithm based on Bayes’ Theorem. It assumes that the features (input variables) are independent of each other given the class label, which is why it is called "naïve". Despite its simple assumptions, it works very well for many complex problems, especially in text classification.

---

## 2. What is Bayes’ Theorem?

**Answer:**  
Bayes' Theorem describes the probability of a class (hypothesis) given some observed data (evidence).  
It is expressed as:  
**P(A|B) = (P(B|A) * P(A)) / P(B)**  
Where:  
- P(A|B) is the probability of hypothesis A given data B  
- P(B|A) is the probability of data B given hypothesis A  
- P(A) is the prior probability of hypothesis A  
- P(B) is the probability of data B

---

## 3. What are the assumptions in Naïve Bayes?

**Answer:**  
- The features are conditionally independent given the class label.
- The contribution of each feature to the probability of a class is independent of other features.

---

## 4. Where is Naïve Bayes commonly used?

**Answer:**  
- Email spam detection  
- Sentiment analysis  
- Document classification  
- Medical diagnosis  
- Recommendation systems

---

## 5. Can Naïve Bayes be used for both binary and multiclass classification?

**Answer:**  
Yes. Naïve Bayes can be applied to binary classification (two classes) as well as multiclass classification problems (more than two classes).

---

## 6. Can you give a simple example of Naïve Bayes classification?

**Answer:**  
Suppose we want to predict if an email is spam or not based on the presence of certain words like "buy", "free", and "offer".  
We calculate the probability of spam given those words using Bayes' Theorem and classify the email accordingly (the class with higher probability).

---

## 7. What are the types of Naïve Bayes classifiers?

**Answer:**  
- **Gaussian Naïve Bayes**: Assumes features follow a normal distribution (used for continuous data).  
- **Multinomial Naïve Bayes**: Used for discrete count data (like word counts in text classification).  
- **Bernoulli Naïve Bayes**: Used for binary/boolean features (like presence or absence of a word).

---

## 8. What is a Confusion Matrix?

**Answer:**  
A confusion matrix is a performance measurement tool for classification models. It shows how many predictions were correct and incorrect by comparing actual and predicted values.

|                | Predicted Positive | Predicted Negative |
|----------------|--------------------|--------------------|
| **Actual Positive** | True Positive (TP)  | False Negative (FN) |
| **Actual Negative** | False Positive (FP) | True Negative (TN)  |

---

## 9. What is Accuracy, Precision, Recall, and F1 Score?

**Answer:**  
- **Accuracy**: Proportion of total correct predictions.  
  **Accuracy = (TP + TN) / (TP + FP + TN + FN)**

- **Precision**: Proportion of correctly predicted positive instances out of all predicted positives.  
  **Precision = TP / (TP + FP)**

- **Recall** (Sensitivity): Proportion of correctly predicted positive instances out of all actual positives.  
  **Recall = TP / (TP + FN)**

- **F1 Score**: Harmonic mean of Precision and Recall.  
  **F1 = 2 * (Precision * Recall) / (Precision + Recall)**

---

## 10. Why is Naïve Bayes considered a fast algorithm?

**Answer:**  
Naïve Bayes is computationally efficient because it calculates probabilities using simple mathematical operations (multiplication and division) and makes strong independence assumptions that simplify the computation, making it faster even on large datasets.
